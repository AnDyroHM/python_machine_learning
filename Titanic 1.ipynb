```python
df = open("python-ml-course/datasets/customer-churn-model/Customer Churn Model.txt","r")
cols = df.readline().strip().split(",")
num_cols = len(cols)
```

```python
counter = 0
main_dic = {}
for col in cols:
    main_dic[col] = []
```

```python
for line in df:
    values = line.strip().split(",")
    for i in range(num_cols):
        main_dic[cols[i]].append(values[i])
    counter+=1
```
